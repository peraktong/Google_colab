```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```
# constant:
g = 9.8
h0=1.2
v0=5.4

def h(t):
  return h0+v0*t-0.5*g*t**2

def v(t):
  return v0-g*t


# at 0.5 second
print("After 0.5 second h=%.3f, v=%.3f"%(h(0.5),v(0.5)))
```

```
t_array = np.linspace(0.1,10,100)
plt.subplot(1,2,1)
plt.plot(t_array,h(t=t_array),"ko",label="h")
plt.xlabel("t")
plt.ylabel("h")
plt.legend()
plt.subplot(1,2,2)
plt.plot(t_array,v(t=t_array),"ko",label="v")
plt.xlabel("t")
plt.ylabel("v")
plt.legend()


fig = matplotlib.pyplot.gcf()

# adjust the size based on the number of visit

fig.set_size_inches(10,10)


```
